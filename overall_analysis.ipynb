```python
import os
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# The path to your CSV files
df_2015 = Path("output/2015_data.csv")
df_2016 = Path("output/2016_data.csv")
df_2017 = Path("output/2017_data.csv")
df_2018 = Path("output/2018_data.csv")
df_2019 = Path("output/2019_data.csv")
#df_2020 = Path("output/2020_data.csv")

# Read the CSV files
df_2015 = pd.read_csv(df_2015)
df_2016 = pd.read_csv(df_2016)
df_2017 = pd.read_csv(df_2017)
df_2018 = pd.read_csv(df_2018)
df_2019 = pd.read_csv(df_2019)
#df_2020 = pd.read_csv(df_2020)

print(df_2015.head())
print(df_2016.head())
print(df_2017.head())
print(df_2018.head())
print(df_2019.head())
#df_2020.head()
```

```python
# Set 'Country' as the index
df_2015.set_index('Country', inplace=True)

# Sort the DataFrame in descending order based on the 'Total' column
df_2015_sorted = df_2015.sort_values(by='Happiness Score', ascending=False)

# Select the columns you want for the stacked bar chart (e.g., 'GDP per Capita', 'Social Support', 'Life Expectancy/Health', etc.)
selected_columns = ['GDP per Capita', 'Social Support', 'Life Expectancy/Health', 'Freedom', 'Trust (government corruption)', 'Generosity']

# Create a stacked bar chart
ax = df_2015[selected_columns].plot(kind='barh', stacked=True, figsize=(15, 20))

# Customize the chart
plt.title('Stacked Bar Chart of Happiness Factors by Country')
plt.xlabel('Countries')
plt.ylabel('Values')
plt.legend(title='Factors', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the chart
plt.tight_layout()
plt.show()
```


#### The stacked bar chart is sorted in descending happiness score

```python
# Select the columns you want for the stacked bar chart (e.g., 'GDP per Capita', 'Social Support', 'Life Expectancy/Health', etc.)
selected_columns = ['GDP per Capita', 'Social Support', 'Life Expectancy/Health', 'Freedom', 'Trust (government corruption)', 'Generosity']

# Create a stacked bar chart
ax = top_10[selected_columns].plot(kind='barh', stacked=True, figsize=(12, 6))

# Customize the chart
plt.title('Stacked Bar Chart of Happiness Factors by Country (Bottom 10)')
plt.xlabel('Values')
plt.ylabel('Countries')
plt.legend(title='Factors', bbox_to_anchor=(1.05, 1), loc='upper left')

# Set the x-axis ticks to be spaced at 0, 1, 2, 3, 4, and 5
#plt.xticks(range(6))

# Add labels to the bars with adjusted positions
for bar in ax.containers:
    ax.bar_label(bar, fmt='%.2f', label_type='center', fontsize=10, color='white', padding=3)

# Show the chart
plt.tight_layout()
plt.show()
```

# Creating a Corellation Coefficient Heatmap

```python
# Calculate the correlation matrix
correlation_matrix = df_2015.corr()

correlation_matrix
```

```python
# Create a heatmap
plt.figure(figsize=(10, 8))
correlation_coefficient_map_2015 = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Coefficient Heatmap')

# Save the figure
plt.savefig('output/correlation_coefficient_map_2015.png')

# Show the figure
plt.show()
```
