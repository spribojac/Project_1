```python
import os
import pandas as pd
```

```python
# Set display options
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# List of file names for the CSV files along with their corresponding years
file_info = [
    ("output/2015_data.csv", "2015"),
    ("output/2016_data.csv", "2016"),
    ("output/2017_data.csv", "2017"),
    ("output/2018_data.csv", "2018"),
    ("output/2019_data.csv", "2019"),
    ("output/2020_data.csv", "2020")
]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through the list of file names and read each file into a DataFrame
for file_name, year in file_info:
    df = pd.read_csv(file_name)
    
    # Add a new column for the year
    df["Year"] = year
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate DataFrames vertically
merged_df = pd.concat(dfs, ignore_index=True)

# Set "Year" as the index
merged_df.set_index("Year", inplace=True)

# Save the merged DataFrame to a new CSV file without the index
merged_file = "output/merged_data.csv"
merged_df.to_csv(merged_file)
```

```python
import pandas as pd

# List of file names for the CSV files along with their corresponding years
file_info = [
    ("output/2015_data.csv", "2015"),
    ("output/2016_data.csv", "2016"),
    ("output/2017_data.csv", "2017"),
    ("output/2018_data.csv", "2018"),
    ("output/2019_data.csv", "2019"),
    ("output/2020_data.csv", "2020")
]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through the list of file names and read each file into a DataFrame
for file_name, year in file_info:
    df = pd.read_csv(file_name)
    
    # Add a new column for the year
    df["Year"] = year
    
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate DataFrames vertically
merged_df = pd.concat(dfs, ignore_index=True)

# Set "Year" as the index
merged_df.set_index("Year", inplace=True)

# Save the merged DataFrame to a new CSV file without the index
merged_file = "output/merged_data.csv"
merged_df.to_csv(merged_file)
```
