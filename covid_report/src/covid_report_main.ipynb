### 1.If GDP has a significant outcome on happiness, will our wealthiest countries have the best handling of Covid-19?
Visualization 1: Relationship between GDP and Happiness
#
The scatter plot illustrates the relationship between the logged GDP per capita (representing the wealth of a country) and the happiness score. Each point on the plot represents a country, and its position is determined by both its GDP and happiness score. If GDP has a significant outcome on happiness, we would expect to see a clear trend or pattern in the plot, indicating that wealthier countries tend to have higher happiness scores.

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatter plot for GDP vs Happiness
plt.figure(figsize=(10, 6))
plt.scatter(df['Logged GDP per capita'], df['Ladder score'], alpha=0.7, label='Data Points')

# Fit a linear regression line
fit = np.polyfit(df['Logged GDP per capita'], df['Ladder score'], 1)
fit_fn = np.poly1d(fit)
plt.plot(df['Logged GDP per capita'], fit_fn(df['Logged GDP per capita']), color='red', label='Linear Fit')

plt.xlabel('Logged GDP per capita')
plt.ylabel('Happiness Score')
plt.title('Relationship between GDP and Happiness Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(df['Logged GDP per capita'], df['Perceptions of corruption'], color='skyblue')
plt.xlabel('Logged GDP per capita')
plt.ylabel('Perceptions of corruption')
plt.title('Relationship between GDP and Perceptions of Corruption')
plt.grid(True)  # Add a grid for better readability
plt.show()

# Calculate correlation
correlation = np.corrcoef(df['Logged GDP per capita'], df['Perceptions of corruption'])[0, 1]

# Define the size of the figure
fig, ax = plt.subplots(figsize=(6, 2))

# Hide axes
ax.axis('off')

# Display the table
table_data = [["Variable 1", "Variable 2", "Correlation"],
              ["Logged GDP per capita", "Perceptions of corruption", f"{correlation:.4f}"]]
ax.table(cellText=table_data, colLabels=None, cellLoc='center', loc='center')

plt.show()
```

The correlation between Logged GDP per capita and Perceptions of corruption is approximately -0.33. This negative correlation suggests that as the GDP per capita increases, perceptions of corruption tend to decrease. In the context of COVID-19 handling, a higher GDP per capita might be associated with better handling of the pandemic, which could contribute to overall happiness in a country.

In summary, countries with a higher GDP per capita tend to have lower levels of perceived corruption, which may positively impact the handling of crises like COVID-19 and subsequently influence overall happiness.

# 2.What about countries with a low perception of Government corruption?
#
We'll first filter the data to include countries with a low perception of government corruption. For this example, let's consider countries with a perception of corruption below 0.3 (you can adjust this threshold as needed).

```python
# Filter data for countries with low perception of corruption
low_corruption_threshold = 0.3
low_corruption_df = df[df['Perceptions of corruption'] < low_corruption_threshold]
```

### Analyze Happiness Score for Low Corruption Countries
Let's analyze the happiness scores for these countries.

```python
# Categorize countries into low, medium, and high perception of corruption
low_corruption_df = df[df['Perceptions of corruption'] < 0.3]
medium_corruption_df = df[(df['Perceptions of corruption'] >= 0.3) & (df['Perceptions of corruption'] < 0.6)]
high_corruption_df = df[df['Perceptions of corruption'] >= 0.6]

# Colors for the boxes
box_colors = ['#FFC3A0', '#FF677D', '#D4A5A5']

# Create a box plot
plt.figure(figsize=(12, 8))
plt.boxplot([low_corruption_df['Ladder score'], medium_corruption_df['Ladder score'], high_corruption_df['Ladder score']],
            labels=['Low Corruption', 'Medium Corruption', 'High Corruption'], patch_artist=True,
            boxprops=dict(facecolor='white', color='black'),
            medianprops=dict(color='black'))

# Fill the boxes with colors
for patch, color in zip(plt.boxplot([low_corruption_df['Ladder score'], medium_corruption_df['Ladder score'],
                                      high_corruption_df['Ladder score']], patch_artist=True)['boxes'], box_colors):
    patch.set_facecolor(color)

plt.xlabel('Perceptions of Corruption')
plt.ylabel('Happiness Score')
plt.title('Happiness Score Distribution based on Perceptions of Corruption')
plt.xticks(fontsize=12, rotation=45)  # Rotate x-axis labels for better visibility
plt.yticks(fontsize=12)
plt.tight_layout()  # Adjust layout for better spacing
plt.grid(True)
plt.show()

from tabulate import tabulate

# Get the number of data points in each category
num_data_points = [len(low_corruption_df), len(medium_corruption_df), len(high_corruption_df)]

# Create a table
table = {
    "Perceptions of Corruption": ['Low Corruption', 'Medium Corruption', 'High Corruption'],
    "Number of Data Points": num_data_points
}

# Print the table
print(tabulate(table, headers="keys", tablefmt="pretty"))
```

```python
plt.figure(figsize=(10,5))

# KDE plot for low corruption
sns.kdeplot(low_corruption_df['Ladder score'], color='red', label='Low Corruption')

# KDE plot for medium corruption
sns.kdeplot(medium_corruption_df['Ladder score'], color='blue', label='Medium Corruption')

# KDE plot for high corruption
sns.kdeplot(high_corruption_df['Ladder score'], color='orange', label='High Corruption')

plt.title('Happiness Score Distribution based on Perceptions of Corruption', size=20)
plt.xlabel('Happiness Score')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

# 3.Does life expectancy have any impact on Covid cases?
We examined the relationship between a country's average life expectancy and the reported number of COVID-19 cases. The scatter plot visually represents this relationship, with life expectancy on the x-axis and COVID cases on the y-axis for each country. The plot allows us to observe any patterns or trends between these variables.

```python
# Scatter plot for Healthy Life Expectancy vs Happiness Score
plt.figure(figsize=(10, 6))
plt.scatter(df['Healthy life expectancy'], df['Ladder score'], alpha=0.7, color='b')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Happiness Score (Ladder score)')
plt.title('Happiness Score vs Healthy Life Expectancy')

# Calculate correlation
correlation = np.corrcoef(df['Healthy life expectancy'], df['Ladder score'])[0, 1]
correlation_label = f'Correlation: {correlation:.2f}'

# Display correlation on the plot
plt.annotate(correlation_label, xy=(0.15, 0.85), xycoords='axes fraction', fontsize=12, color='r')

plt.grid(True)
plt.show()

# Get the top 8 countries for Healthy Life Expectancy and Happiness Score
top_countries_health = df.nlargest(8, 'Healthy life expectancy')
top_countries_happiness = df.nlargest(8, 'Ladder score')

variables = ['Healthy life expectancy', 'Ladder score']
top_countries = [top_countries_health, top_countries_happiness]

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.35
index = np.arange(len(top_countries[0]))

for i in range(len(variables)):
    bars = ax.bar(index + i * bar_width, top_countries[i][variables[i]], bar_width, label=variables[i])
    for bar in bars:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
                f'{bar.get_height():.2f}', ha='center', va='bottom')

ax.set_xlabel('Country')
ax.set_ylabel('Values')
ax.set_title('Top 8 Countries: Healthy Life Expectancy and Happiness Score')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(top_countries[0]['Country name'])
ax.legend()
plt.tight_layout()
plt.show()
```
